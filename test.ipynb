```python
df = pd.read_csv('BankChurners.csv')
df.head()
```

```python
# Transformation function
def sigma(value):
    if value < 20:
        return "YOUNG"
    elif value < 30:
        return "ADULT"
    elif value < 50:
        return "SENIOR"
    elif value < 65:
        return "OLD"
    else:
        return "NOT_WORKER"
```

```python
df["Age Class"] = df['Customer_Age'].apply(sigma)
df_filtered = df[df['Education_Level'] == 'Graduate']
df_target = pd.DataFrame()
df_target['CLIENTNUM'] = df_filtered['CLIENTNUM']
df_target['targetMatch'] = 1
behaviors = ['Marital_Status', 'Attrition_Flag', 'Age Class', 'Income_Category', 'Income_Category']
cc_graduate = CC(df, df_target, behaviors, 'Graduate', 'CLIENTNUM', 'targetMatch')
cc_graduate.analyse()
_, analysis = cc_graduate.getAnalysis()
analysis
```

```python
df["Age Class"] = df['Customer_Age'].apply(sigma)
df_filtered = df[df['Education_Level'] == 'High School']
df_target = pd.DataFrame()
df_target['CLIENTNUM'] = df_filtered['CLIENTNUM']
df_target['targetMatch'] = 1
behaviors = ['Marital_Status', 'Attrition_Flag', 'Age Class', 'Income_Category', 'Income_Category']
cc_highschool = CC(df, df_target, behaviors, 'High School', 'CLIENTNUM', 'targetMatch')
cc_highschool.analyse()
_, analysis = cc_highschool.getAnalysis()
analysis
```
